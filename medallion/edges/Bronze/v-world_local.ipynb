```python
soil_type_data = gpd.read_file(r"C:\Users\bagg8\Downloads\ASIT_DEEPSOIL_QLT_AREA\ASIT_DEEPSOIL_QLT_AREA.shp")
soil_type_data = soil_type_data.set_crs(epsg=5181).to_crs(epsg=4326)

boundary_4326 = gangnam3.to_crs(epsg=4326)

# 1차 clip 시도 — 좌표계 맞추고 결과 확인
result = gpd.clip(soil_type_data, boundary_4326)
print(f"clip 결과: {len(result)}건")
result.head()
```

```python
# 위 clip 시도가 0건 나오는 경우 — CRS를 강제로 덮어써서 맞춤
# sig.shp는 5179, V-World SHP는 5181로 고정 후 5181 기준으로 clip
def get_seoul_data_final(national_data, boundary):
    boundary.set_crs(epsg=5179, allow_override=True, inplace=True)
    if national_data.crs is None or national_data.crs.to_epsg() != 5181:
        national_data.set_crs(epsg=5181, allow_override=True, inplace=True)

    target_boundary = boundary.to_crs(epsg=5181)
    return gpd.clip(national_data, target_boundary)

sig_data = gpd.read_file(r"C:\Users\bagg8\Downloads\sig\sig\sig.shp", encoding='cp949')
seoul_boundary = sig_data[sig_data['SIG_CD'].str.startswith('11')]
gangnam3 = sig_data[sig_data['SIG_KOR_NM'].isin(['강남구', '송파구', '강동구'])]

seoul_soil_type      = get_seoul_data_final(soil_type_data,      gangnam3)
seoul_gravel         = get_seoul_data_final(gravel_data,         gangnam3)
seoul_soil_depth     = get_seoul_data_final(soil_depth_data,     gangnam3)
seoul_drainage_class = get_seoul_data_final(drainage_class_data, gangnam3)

print(f"심토토성: {len(seoul_soil_type)}건")
print(f"자갈함량: {len(seoul_gravel)}건")
print(f"유효토심: {len(seoul_soil_depth)}건")
print(f"배수등급: {len(seoul_drainage_class)}건")
```

```python
# 강남·송파·강동구 필터링된 SHP 저장 — 이후 bronze_raw에서 Blob Storage로 업로드
seoul_gravel.to_file('./data/seoul/자갈함량/gangnam_songpa_gangdong_gravel.shp', encoding='utf-8')
seoul_soil_type.to_file('./data/seoul/심토토성/gangnam_songpa_gangdong_soil_type.shp', encoding='utf-8')
seoul_soil_depth.to_file('./data/seoul/유효토심/gangnam_songpa_gangdong_soil_depth.shp', encoding='utf-8')
seoul_drainage_class.to_file('./data/seoul/배수등급/gangnam_songpa_gangdong_drainage_class.shp', encoding='utf-8')
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
plt.rcParams['axes.unicode_minus'] = False

gravel    = gpd.read_file('./data/seoul/자갈함량/gangnam_songpa_gangdong_gravel.shp')
soil_type = gpd.read_file('./data/seoul/심토토성/gangnam_songpa_gangdong_soil_type.shp')
soil_depth= gpd.read_file('./data/seoul/유효토심/gangnam_songpa_gangdong_soil_depth.shp')
drainage  = gpd.read_file('./data/seoul/배수등급/gangnam_songpa_gangdong_drainage_class.shp')

fig, axes = plt.subplots(2, 2, figsize=(20, 18))

gravel.plot(ax=axes[0, 0], column='SUR_STON', cmap='YlOrBr', legend=True,
            edgecolor='black', linewidth=0.1)
axes[0, 0].set_title('자갈함량 (%)', fontsize=16, fontweight='bold')

soil_type.plot(ax=axes[0, 1], column='DEEPSOIL', cmap='Set3', legend=True,
               edgecolor='black', linewidth=0.1)
axes[0, 1].set_title('심토토성', fontsize=16, fontweight='bold')

soil_depth.plot(ax=axes[1, 0], column='VLDSOILDEP', cmap='Greens', legend=True,
                edgecolor='black', linewidth=0.1)
axes[1, 0].set_title('유효토심 (cm)', fontsize=16, fontweight='bold')

drainage.plot(ax=axes[1, 1], column='SOILDRA', cmap='RdYlBu', legend=True,
              edgecolor='black', linewidth=0.1)
axes[1, 1].set_title('배수등급', fontsize=16, fontweight='bold')

for ax in axes.flat:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
